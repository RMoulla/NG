````python
import pandas as pd
dataset=pd.read_csv('products.csv')
dataset.head()
````

````python
!pip install -q gradio sentence-transformers

import pandas as pd
import numpy as np
import gradio as gr
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

# Chargement des données
df = pd.read_csv("products.csv")

# Nettoyage minimal
df = df.fillna("")

# Texte utilisé pour la recherche sémantique
df["search_text"] = (
    df["title"].astype(str) + " " +
    df["brand"].astype(str) + " " +
    df["category"].astype(str) + " " +
    df["sub_category"].astype(str) + " " +
    df["description"].astype(str)
)

# Modèle d'embeddings multilingue
model = SentenceTransformer("sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2")

# Encodage des produits
product_embeddings = model.encode(
    df["search_text"].tolist(),
    convert_to_numpy=True,
    show_progress_bar=True
)

def search_products(query, top_k=8):
    if not query.strip():
        return "<p>Veuillez saisir une recherche.</p>"

    query_embedding = model.encode([query], convert_to_numpy=True)
    scores = cosine_similarity(query_embedding, product_embeddings)[0]

    df_results = df.copy()
    df_results["score"] = scores
    df_results = df_results.sort_values("score", ascending=False).head(top_k)

    if df_results["score"].max() < 0.15:
        return "<p>Aucun produit suffisamment pertinent n’a été trouvé.</p>"

    html = """
    <div class="results-grid">
    """

    for _, row in df_results.iterrows():
        title = row["title"]
        brand = row["brand"]
        price = row["selling_price"]
        rating = row["average_rating"]
        image = row["images"]
        description = str(row["description"])[:160] + "..."

        html += f"""
        <div class="product-card">
            <img src="{image}" class="product-img">
            <div class="product-info">
                <h3>{title}</h3>
                <p class="brand">{brand}</p>
                <p class="desc">{description}</p>
                <div class="meta">
                    <span class="price">{price} €</span>
                    <span class="rating">⭐ {rating}</span>
                </div>
            </div>
        </div>
        """

    html += "</div>"
    return html

css = """
.results-grid {
    display: grid;
    grid-template-columns: repeat(auto-fill, minmax(220px, 1fr));
    gap: 18px;
    margin-top: 20px;
}

.product-card {
    border: 1px solid #e5e5e5;
    border-radius: 16px;
    padding: 14px;
    background: white;
    box-shadow: 0 4px 14px rgba(0,0,0,0.06);
    transition: transform 0.2s ease;
}

.product-card:hover {
    transform: translateY(-4px);
}

.product-img {
    width: 100%;
    height: 170px;
    object-fit: contain;
    border-radius: 12px;
    background: #f7f7f7;
}

.product-info h3 {
    font-size: 16px;
    margin: 12px 0 4px;
}

.brand {
    color: #666;
    font-size: 13px;
}

.desc {
    font-size: 13px;
    color: #444;
    min-height: 55px;
}

.meta {
    display: flex;
    justify-content: space-between;
    margin-top: 12px;
    font-weight: bold;
}

.price {
    color: #111;
}

.rating {
    color: #f5a623;
}
"""

with gr.Blocks(css=css) as demo:
    gr.HTML("""
    <div style="text-align:center; margin-bottom:30px;">
        <h1>Moteur de recherche sémantique</h1>
        <p>Recherchez des vêtements, chaussures ou accessoires avec une phrase naturelle.</p>
    </div>
    """)

    query = gr.Textbox(
        label="Votre recherche",
        placeholder="Exemple : baskets confortables pour marcher toute la journée"
    )

    button = gr.Button("Rechercher")
    output = gr.HTML()

    button.click(
        fn=search_products,
        inputs=query,
        outputs=output
    )

demo.launch(share=True)
````
